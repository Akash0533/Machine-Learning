```python
df1=df.drop(['Id'],axis=1)
print(df1.head(2))
m=df1.mean()
print(m)
print(m.index)
print(m.values)
```

```python
plt.bar(m.index,m.values,color='red')
plt.grid()
plt.xlabel('Sepal Patel length&Width')
plt.ylabel('Mean')
plt.title('Mean of Sepal/Petal Lenth and Width')
plt.show()
```

```python
plt.plot(m.index,m.values,color='red')
plt.grid()
plt.xlabel('Sepal Patel length&Width')
plt.ylabel('Mean')
plt.title('Mean of Sepal/Petal Lenth and Width')
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(6,5))
sns.barplot(x=m.index,y=m.values)
plt.grid()
plt.title('Bar plot (mean)')
plt.xlabel('Petal,Sapel Length&Width')
plt.ylabel('mean')
plt.show
```

```python
import seaborn as sns
plt.figure(figsize=(6,5))
sns.lineplot(x=m.index,y=m.values)
plt.grid()
plt.title('Bar plot (mean)')
plt.xlabel('Petal,Sapel Length&Width')
plt.ylabel('mean')
plt.show
```

# Scatter Plot

```python
plt.scatter(x=df[['PetalLengthCm']],y=df[['PetalWidthCm']])
plt.grid()
plt.title('Iris Flowers leaves Lenth & Width')
plt.xlabel('PetalLengthCm')
plt.ylabel('PetalWidthCm')
plt.show()
```

```python
sns.scatterplot(data=df,x=('PetalLengthCm'),y=('PetalWidthCm'),hue='Species')
plt.grid()
plt.title('Iris Flowers leaves Lenth & Width')
plt.xlabel('PetalLengthCm')
plt.ylabel('PetalWidthCm')
plt.show()
```
